```python
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

### Лабораторная №1
Решить задачу Коши методом Эйлера и методом Эйлера с пересчетом со значениями шага $h_1=0.1$ и $h_2 = 0.01$. Сравнить результаты графически.  

**2 вариант**:  
$$y'=-\frac{y}{2x} + x^2$$
$$y(1)=1$$
$$x \in [1,2]$$

```python
def y_diff(x, y): return -y/(2*x) + x * x


x_0, y_0, start, end = 1, 1, 1, 2
h = [0.1, 0.01]
```

#### Метод Эйлера и метод Эйлера с перерасчетом

```python
def euler(y_diff, x: float, y: float, start: float, end: float, h: float, recalculation: bool = False) -> list:
    Y = [y]
    if recalculation == True:
        k = 1
    else:
        k = 0

    for x_i in np.arange(start, end, h):
        
        Y.append(Y[-1] + h/(1+k) * (y_diff(x_i, Y[-1]) + k * y_diff(x_i + h, Y[-1] + h * y_diff(x_i, Y[-1]))))
    return Y
```

**Note:**  
- Видоизменение формулы. Сделали все под одну
$$
y_{i+1} = y_i + \frac{h}{1 + k}(f(x_i, y_i) + k * f(x_{i+1}, y_{i+1})), \\
\text{где } k \in 0, 1
$$

#### Метод Эйлера

```python
# Решение с шагом 0.1
y_euler_h0 = np.array(euler(y_diff, x_0, y_0, start, end, h[0]))
x_euler_h0 = np.array([i for i in np.arange(start, end + h[0], h[0])])

# Решение с шагом 0.01
y_euler_h1 = euler(y_diff, x_0, y_0, start, end, h[1])
x_euler_h1 = np.linspace(start, end, len(y_euler_h1))
```

#### Метод Эйлера с пересчетом

```python
# Решение с шагом 0.1
y_euler_recalc_h0 = np.array(euler(y_diff, x_0, y_0, start, end, h[0], True))
x_euler_recalc_h0 = np.linspace(start, end, len(y_euler_recalc_h0))

# Решение с шагом 0.01
y_euler_recalc_h1 = euler(y_diff, x_0, y_0, start, end, h[1], True)
x_euler_racalc_h0 = np.linspace(start, end, len(y_euler_recalc_h1))
```

#### Построение графиков

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=x_euler_h0, y=y_euler_h0, mode='lines', name='e_h<sub>0</sub>=0.1'))
fig.add_trace(go.Scatter(x=x_euler_h1, y=y_euler_h1, mode='lines', name='e_h<sub>0</sub>=0.01'))

fig.add_trace(go.Scatter(x=x_euler_recalc_h0, y=y_euler_recalc_h0, mode='lines', name='erec_h<sub>1</sub>=0.1'))
fig.add_trace(go.Scatter(x=x_euler_racalc_h0, y=y_euler_recalc_h1, mode='lines', name='erec_h<sub>1</sub>=0.01'))
fig.update_layout(legend_orientation='h',
                  legend = dict(x=.5, xanchor='center'),
                  title='Метод Эйлера и метод Эйлера с пересчетом')

fig.show()
```
